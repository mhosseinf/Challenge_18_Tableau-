```python
from pathlib import Path
import pandas as pd
import pprint
import os
```

```python
csv_stations = Path("../Resources/202212-citibike-tripdata.csv")
citibike_df1 = pd.read_csv(csv_stations,low_memory=False)
citibike_df1['started_at'] = pd.to_datetime(citibike_df1['started_at'])
citibike_df1['ended_at'] = pd.to_datetime(citibike_df1['ended_at'])
citibike_df1['start_date'] = citibike_df1['started_at'].dt.date
citibike_df1['start_time'] = citibike_df1['started_at'].dt.time
citibike_df1['ended_date'] = citibike_df1['ended_at'].dt.date
citibike_df1['ended_time'] = citibike_df1['ended_at'].dt.time
citibike_df1.head()
```

```python
csv_stations = Path("../Resources/202301-citibike-tripdata.csv")
citibike_df2 = pd.read_csv(csv_stations,low_memory=False)
citibike_df2['started_at'] = pd.to_datetime(citibike_df2['started_at'])
citibike_df2['ended_at'] = pd.to_datetime(citibike_df2['ended_at'])
citibike_df2['start_date'] = citibike_df2['started_at'].dt.date
citibike_df2['start_time'] = citibike_df2['started_at'].dt.time
citibike_df2['ended_date'] = citibike_df2['ended_at'].dt.date
citibike_df2['ended_time'] = citibike_df2['ended_at'].dt.time
citibike_df2.head()
```

```python
csv_stations = Path("../Resources/202302-citibike-tripdata.csv")
citibike_df3 = pd.read_csv(csv_stations,low_memory=False)
citibike_df3['started_at'] = pd.to_datetime(citibike_df3['started_at'])
citibike_df3['ended_at'] = pd.to_datetime(citibike_df3['ended_at'])
citibike_df3['start_date'] = citibike_df3['started_at'].dt.date
citibike_df3['start_time'] = citibike_df3['started_at'].dt.time
citibike_df3['ended_date'] = citibike_df3['ended_at'].dt.date
citibike_df3['ended_time'] = citibike_df3['ended_at'].dt.time
citibike_df3.head()
```

```python
df1_subset = citibike_df1.head(270000)
df2_subset = citibike_df2.head(500000)
df3_subset = citibike_df3.head(230000)
```

```python
result_df = pd.concat([df1_subset, df2_subset, df3_subset], axis=0)
result_df
```

```python
result_df.reset_index(drop=True, inplace=True)
result_df
```

```python
result_df.to_csv("../Resources/citibike-tripdata-merged.csv", index=False)
```

```python
# Convert 'start_date' column to datetime format
result_df['start_date'] = pd.to_datetime(result_df['start_date'])

# Calculate total ridership for each period
total_ridership_dec = len(result_df[result_df['start_date'] < '2023-01-01'])
total_ridership_jan = len(result_df[(result_df['start_date'] >= '2023-01-01') & (result_df['start_date'] < '2023-02-01')])
total_ridership_feb = len(result_df[result_df['start_date'] >= '2023-02-01'])

# Find the difference
ridership_growth1 = ((total_ridership_jan - total_ridership_dec) / total_ridership_dec) * 100
ridership_growth2 = ((total_ridership_feb - total_ridership_jan) / total_ridership_jan) * 100

print(f'Percentage growth from December to January: {ridership_growth1:.2f}%')
print(f'Percentage growth from January to February: {ridership_growth2:.2f}%')
```
